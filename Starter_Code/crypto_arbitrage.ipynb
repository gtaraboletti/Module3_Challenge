###  Import the required libraries and dependencies.

```python
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
# imports the first csv file (bitstamp exchange data in this case) and sets parameters 
bitstamp = pd.read_csv(
    Path("../Starter_Code/Resources/bitstamp.csv"),
    index_col="Timestamp",
    parse_dates=True,
    infer_datetime_format=True)
```

```python
# arbitrage_spread of the different periods
arbitrage_spread_early = coinbase_sliced.loc["2018-01-01"] -  bitstamp_sliced.loc["2018-01-01"]
arbitrage_spread_mid = coinbase_sliced.loc["2018-02-24"] -  bitstamp_sliced.loc["2018-02-24"]
arbitrage_spread_late = coinbase_sliced.loc["2018-03-14"] -  bitstamp_sliced.loc["2018-03-14"]
```

```python
# conditionals to find distances greater than 0
boofilter_early= arbitrage_spread_early > 0
boofilter_mid = arbitrage_spread_mid > 0
boofilter_late = arbitrage_spread_late > 0
```

```python
#display summary statistics of arbitrage data greater than 0 (Jan 1, 2018)
arbitrage_spread_early.loc[boofilter_early].describe()
```

```python
#display summary statistics of arbitrage data greater than 0 (Feb 24, 2018)
arbitrage_spread_mid.loc[boofilter_mid].describe()
```

```python
#display summary statistics of arbitrage data greater than 0 (Mar 14, 2018)
arbitrage_spread_late.loc[boofilter_late].describe()
```

```python
#Jan 1, 2018: calculates spread returns 
spread_return_early = arbitrage_spread_early.loc[boofilter_early] /  bitstamp_sliced.loc["2018-01-01"] 
spread_return_early
```

```python
#Feb 24, 2018: calculates spread returns 
spread_return_mid = arbitrage_spread_mid.loc[boofilter_mid] / bitstamp_sliced.loc["2018-02-24"] 
spread_return_mid
```

```python
#Mar 14, 2018: calculates spread returns 
spread_return_late = arbitrage_spread_late.loc[boofilter_late] / bitstamp_sliced.loc["2018-03-14"]
spread_return_late
```

```python
#Pull trades with returns of greater than 1%
profitable_trades_early = spread_return_early[spread_return_early >= 0.01]

# Review the first five profitable trades
profitable_trades_early.head()
```

```python
#Pull trades with returns of greater than 1%
profitable_trades_mid = spread_return_mid[spread_return_mid >= 0.01]
profitable_trades_mid.head()
```

```python
#Pull trades with returns of greater than 1%
profitable_trades_late = spread_return_late[spread_return_late >= 0.01]
profitable_trades_late.head()
```

```python
# summary statistics of profitable trades Jan 1, 2018
profitable_trades_early.describe()
```

```python
# summary statistics of profitable trades Feb 24, 2018
profitable_trades_mid.describe()
```

```python
# summary statistics of profitable trades Mar 14, 2018
profitable_trades_late.describe()
```

```python
# Calculates potential profits, drops "NaN", and shows summary statistics Jan 1, 2018 
profit_early = profitable_trades_early * bitstamp_sliced.loc["2018-01-01"]
profit_per_trade_early = profit_early.dropna()
profit_per_trade_early.describe()
```

**Answer:** A trend I noticed is that the closing price for Bitcoin across both exchanges tends to be lower at the end of the day than it is at the beginning of the day. The closing price of Bitcoin between Jan. 2018 and Apr. 2018 decreased overall across both exchanges, suggesting that Bitcoin is decreasing in value. This may suggest a good time to buy Bitcoin, so long as there is evidence that Bitcoin may turn around rather than drop completely. The profits per trade dip at the end of the day. Also, cumulative profits rise sharply at the end of the day. In attempting to find any 1% difference between the two exchanges at the end of the timeframe of the data, I had to look through many days just to find a date with even 1 data point that was greater than 1%.
